P-2.33 Write a Python program that inputs a polynomial in standard algebraic notation and outputs the first derivative of that polynomial.

```python
class StandardPolynomial:
  def __init__(self, deg):
    self.deg = deg

  def get_poly(self):
    poly = []
    for i, w in zip(range(self.deg,-1,-1),range(self.deg + 1)):
      term = int(input("Enter Coefficient for Variable Exponent " + str(i) + ": "))
      poly.insert(0, term)
    return poly

  def get_first(self, poly):
    deriv = []
    for q in range(self.deg, -1, -1):
      coeff = q * poly[q]
      deriv.insert(0, coeff)
    return deriv

  def show_giv(self, list):
    giv = ""
    lead = len(list) - 1
    for i in range(lead, -1, -1):
      term = list[i]
      if i == 1:
        giv += str(term) +'x' + ' + '
      elif i == 0:
        giv += str(term) + ' = 0'
      else:
        giv += str(term) +'x^' + str(i) + ' + '
    return giv

  def show_res(self, list):
    res = ""
    lead = len(list) - 1
    for i in range(lead, -1, -1):
      term = list[i]
      if (i-1) == 1:
        res += str(term) +'x' + ' + '
      elif (i-1) == 0:
        res += str(term)
      elif (i-1) == -1:
        res += ' = 0'
      else:
        res += str(term) +'x^' + str(i-1) + ' + '
    return res

def main():
  ques = StandardPolynomial(int(input('Enter degree: ')))
  given = ques.get_poly()
  answer = ques.get_first(given)

  print("Given: " + ques.show_giv(given))
  print("Result: " + ques.show_res(answer))

main()
```

P-2.39 Develop an inheritance hierarchy based upon a Polygon class that has abstract methods area( ) and perimeter( ). Implement classes Triangle, Quadrilateral, Pentagon, Hexagon, and Octagon that extend this base class, with the obvious meanings for the area( ) and perimeter( ) methods. Also implement classes, IsoscelesTriangle, EquilateralTriangle, Rectangle, and Square, that have the appropriate inheritance relationships.
```python
import math

class Polygon:
    def __init__(self, vertices):
        self.vertices = vertices

    def area(self):
        raise NotImplementedError("Area calculation not implemented for this polygon")

    def perimeter(self):
        raise NotImplementedError("Perimeter calculation not implemented for this polygon")


class Triangle(Polygon):
    def area(self):
        a, b, c = self.vertices
        s = (a + b + c) / 2
        return math.sqrt(s * (s - a) * (s - b) * (s - c))

    def perimeter(self):
        return sum(self.vertices)


class Quadrilateral(Polygon):
    def area(self):
        raise NotImplementedError("Area calculation for a generic quadrilateral not implemented")

    def perimeter(self):
        return sum(self.vertices)


class Rectangle(Quadrilateral):
    def area(self):
        return self.vertices[0] * self.vertices[1]  # Length * Width


class Square(Rectangle):
    def __init__(self, side):
        super().__init__([side, side])  # Initialize as a rectangle with equal sides


class RegularPolygon(Polygon):
    def __init__(self, num_sides, side_length):
        self.num_sides = num_sides
        self.side_length = side_length
        self.vertices = [side_length] * num_sides

    def perimeter(self):
        return self.num_sides * self.side_length


class Pentagon(RegularPolygon):
    def __init__(self, side_length):
        super().__init__(5, side_length)

    def area(self):
        return (1/4) * math.sqrt(5 * (5 + 2 * math.sqrt(5))) * self.side_length**2


class Hexagon(RegularPolygon):
    def __init__(self, side_length):
        super().__init__(6, side_length)

    def area(self):
        return (3 * math.sqrt(3) / 2) * self.side_length**2


class Octagon(RegularPolygon):
    def __init__(self, side_length):
        super().__init__(8, side_length)

    def area(self):
        return 2 * (1 + math.sqrt(2)) * self.side_length**2


class IsoscelesTriangle(Triangle):
    def __init__(self, equal_side, base):
        super().__init__([equal_side, equal_side, base])


class EquilateralTriangle(IsoscelesTriangle):
    def __init__(self, side_length):
        super().__init__(side_length, side_length)

    def area(self):
        return (math.sqrt(3) / 4) * self.vertices[0]**2


def main():
    while True:
        print("Choose a polygon type:")
        print("1. Triangle")
        print("2. Quadrilateral")
        print("3. Pentagon")
        print("4. Hexagon")
        print("5. Octagon")
        print("6. Isosceles Triangle")
        print("7. Equilateral Triangle")
        print("8. Rectangle")
        print("9. Square")
        print("10. Exit")

        choice = int(input("Enter your choice: "))
        polygon = None

        if choice == 10:
            print("Exiting the program.")
            break

        if choice == 1:
            vertices = [float(input(f"Enter side {i + 1}: ")) for i in range(3)]
            polygon = Triangle(vertices)
        elif choice == 2:
            print("Please enter the lengths of the sides:")
            vertices = [float(input(f"Enter side {i + 1}: ")) for i in range(4)]
            polygon = Quadrilateral(vertices)
        elif choice == 3:
            side_length = float(input("Enter side length: "))
            polygon = Pentagon(side_length)
        elif choice == 4:
            side_length = float(input("Enter side length: "))
            polygon = Hexagon(side_length)
        elif choice == 5:
            side_length = float(input("Enter side length: "))
            polygon = Octagon(side_length)
        elif choice == 6:
            equal_side = float(input("Enter equal side: "))
            base = float(input("Enter base: "))
            polygon = IsoscelesTriangle(equal_side, base)
        elif choice == 7:
            side_length = float(input("Enter equal side: "))
            polygon = EquilateralTriangle(side_length)
        elif choice == 8:
            length = float(input("Enter length: "))
            width = float(input("Enter width: "))
            polygon = Rectangle([length, width])
        elif choice == 9:
            side = float(input("Enter side length: "))
            polygon = Square(side)

        if polygon is not None:
            print("Area:", polygon.area())
            print("Perimeter:", polygon.perimeter())
        else:
            print("Invalid choice. Please try again.")

        # Ask if the user wants to run the program again
        run_again = input("Do you want to calculate another polygon? (yes/no): ").strip().lower()
        if run_again != "yes":
            print("Exiting the program.")
            break

if __name__ == "__main__":
    main()
```
