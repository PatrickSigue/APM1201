```python
import random as r
import time as t
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

list_size = [100, 500, 1000, 10000, 25000, 500000, 1000000]
q = []


def test():
  for i in list_size:
    S =[]
    for w in range(i):
      S.append(r.randint(0, 10000))
    start_time = t.time()
    S.sort()
    end_time = t.time()
    q.append(end_time - start_time)
  return q

execution_times = test()

df = pd.DataFrame({'List Size': list_size, 'Execution Time': execution_times})

plt.figure(figsize=(10, 6))
sns.lineplot(x='List Size', y='Execution Time', data=df)

plt.title('Execution Time vs List Size')
plt.xlabel('List Size')
plt.ylabel('Execution Time (seconds)')

plt.grid(True)
plt.xscale('log')
plt.yscale('log')

plt.show()
```
